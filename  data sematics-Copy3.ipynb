```python
#import libraries
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('TRAINING.csv')
```

```python
df.head()
```

### data preprocessing and data cleaning 

```python
#finding the null values
df.isnull().sum()
```

```python
#mappping the values in attributes using one hot coding or you can use ordinal encoding 
f={'yes':1,'no':0,'YES':1,'NO':0}
```

```python
df['roof']=df['roof'].map(f)
```

```python
df.fillna(df.median(),inplace=True)
```

```python
#import category_encoders as ce
```

```python
#dfy=ce.OrdinalEncoder(cols=['roof'])
#dfa=dfy.fit_transform(df,['outcome'])
#dfa
 
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.describe(include='object')
```

```python
df.roof.value_counts() / len(df.roof)
```

```python
df.roof.value_counts(normalize=True)
```

```python
df.roof.unique()
```

```python
#here we have removed $ symbol for the price attribute
df['Price'] = df['Price'].replace({'\$': '', ',': ''}, regex=True)
```

```python
#convert to float for price
df['Price'] = df['Price'].astype(float)
```

```python
df
```

```python
#split the data into x and y
x=df.drop('Grade',axis=1)
y=df['Grade']
```

```python
x.head(3)
```

```python
x.shape
```

### xgboost model

```python
from xgboost import XGBClassifier
```

```python
 xgb = XGBClassifier()
```

```python
 xgb
```

```python
xgb.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=22)
```

```python
xgb.fit(x_train, y_train)
```

```python
y_test_pred = xgb.predict(x_test)
```

```python
from sklearn import metrics
```

```python
accuracy_before=metrics.accuracy_score(y_test,y_test_pred)
accuracy_before
```

```python
 from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgb = XGBClassifier( n_estimators=100)
```

```python
 param_grid= { "n_estimators"      : [100,1000],    
               "learning_rate"   : [0.1,0.001,0.2,0.3],             
               "nthread"       : [4,5,6,7,8,2],  
               "subsample" : [1.0],
               "max_depth"        : [6,8,9,5],
             "colsample_bytree"       :[0.5,0.8],
             "min_child_weight"  : [3,4],
             "reg_alpha"  : [0.03],
             "reg_lambda" : [1],
            }
```

```python
 optimized_tree = RandomizedSearchCV(xgb, param_grid, scoring = 'accuracy', verbose=2, cv=5) 
```

```python
 optimized_tree.fit(x_train, y_train)
```

```python
 optimized_tree.best_params_
```

```python
 y_test_pred=optimized_tree.predict(x_test)
```

### accuracy after tuning 

```python
 metrics.accuracy_score(y_test, y_test_pred)
```
